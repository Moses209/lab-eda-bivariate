```python
#Part 1: Analyzing Best-Seller Trends Across Product Categories
#Objective: Understand the relationship between product categories and their best-seller status.
#Crosstab Analysis:
#1.1 Create a crosstab between the product category and the isBestSeller status.
#1.2 Are there categories where being a best-seller is more prevalent?

crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])

crosstab_result
```

```python
crosstab_result.head()
```

```python
#1.2 Are there categories where being a best-seller is more prevalent?
#Yes, some categories do show a higher prevalence of best-sellers — meaning a greater proportion of products in those categories
#are marked as isBestSeller = True.
```

```python
from scipy.stats import chi2_contingency

#Statistical Tests:
#2.1 Conduct a Chi-square test to determine if the best-seller distribution is independent of the product category.

# Chi-square test for 'category' and 'isBestSeller'
chi2_statistic, chi2_p_value, _, _ = chi2_contingency(crosstab_result)

chi2_statistic, chi2_p_value
```

```python
from scipy.stats.contingency import association
#2.2 Compute Cramér's V to understand the strength of association between best-seller status and category.
association(crosstab_result, method="cramer")
```

```python
#The Cramér's V value of 0.148 suggests a weak association between the two variables. Though statistically significant, 
#the strength of this relationship is not very strong in practical terms.
```

```python
#Visualizations:
#3 Visualize the relationship between product categories and the best-seller status using a stacked bar chart.

# Plotting a stacked bar chart for the 'crosstab_result' data
crosstab_result.head().plot(kind="bar", stacked=True)
plt.show()
```

```python
#Part 2: Exploring Product Prices and Ratings Across Categories and Brands
# Objective: Investigate how different product categories influence product prices.

#0. Preliminary Step: Remove outliers in product prices.

#For this purpose, we can use the IQR (Interquartile Range) method. Products priced below the first quartile minus 1.5 
#times the IQR or above the third quartile plus 1.5 times the IQR will be considered outliers and removed from the dataset. 
#The next steps will be done with the dataframe without outliers.

#Hint: you can check the last Check For Understanding at the end of the 
#lesson EDA Bivariate Analysis for a hint on how to do this.
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
    
    # Define bounds for the outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
    
    # Identify the outliers
df_filter = df[(df['price'] > lower_bound) & (df['price'] < upper_bound)]
```

```python
#1 Violin Plots:
#1.1 Use a violin plot to visualize the distribution of price across different product categories. 
#Filter out the top 20 categories based on count for better visualization.
# Plotting a violin plot to visualize the distribution of 'SalePrice' for each 'MSZoning' category, using the "coolwarm" color palette
sns.violinplot(data=df, x='category', y='price', palette="coolwarm")
plt.show()
```

```python
#1 Violin Plots:
#1.1 Use a violin plot to visualize the distribution of price across different product categories. 
#Filter out the top 20 categories based on count for better visualization.
# Filter top 20 categories by count

top_20_categories = df['category'].value_counts().nlargest(20).index
df_top20 = df[df['category'].isin(top_20_categories)]

sns.violinplot(data=df_top20, x='category', y='price', palette="coolwarm")

plt.xticks(rotation = 90)
plt.show()
```

```python
#1.2 Which product category tends to have the highest median price? Don't filter here by top categories.
```

```python
#Bar Charts:

#2.1 Create a bar chart comparing the average price of products for the top 10 product categories (based on count).

# Identify top 10 product categories by count
top_10_categories = df['category'].value_counts().nlargest(10).index

# Filter the DataFrame for those top 10 categories
df_top10 = df[df['category'].isin(top_10_categories)]

# Calculate average price per category
avg_price_per_category = df_top10.groupby('category')['price'].mean().sort_values(ascending=False)

sns.barplot(x=avg_price_per_category.index, y=avg_price_per_category.values, palette="coolwarm")
plt.xticks(rotation = 90)
plt.show()
```

```python
#2.2 Which product category commands the highest average price? Don't filter here by top categories.
```

```python
# Box Plots:

#3.1 Visualize the distribution of product ratings based on their category using side-by-side box plots. 
#Filter out the top 10 categories based on count for better visualization.

sns.boxplot(data=df_top10, x='category', y='stars', palette="coolwarm")
plt.xticks(rotation = 90)
plt.show()
```

```python
#3.2 Which category tends to receive the highest median rating from customers? Don't filter here by top categories.
```

```python
#Part 3: Investigating the Interplay Between Product Prices and Ratings
#Objective: Analyze how product ratings (stars) correlate with product prices.

#Correlation Coefficients:

#1.1 Calculate the Pearson correlation coefficient between price and stars.

correlation = df['price'].corr(df['stars'])
correlation
```

```python
#1.2 Is there a significant correlation between product price and its rating?
#The correlation is negative, suggesting that as price increases, ratings slightly decrease.
#The value is very close to 0, indicating a very weak linear relationship.
```

```python
#1.1.1  Calculate the Spearman correlation coefficient between price and stars.

correlation = df['price'].corr(df['stars'], method='spearman')
correlation
```
